AREA PLOTS, HISTOGRAMS AND BAR PLOTS

```python
#import the modules
import numpy as np
import pandas as pd
```

```python
#downoad and read the data
df_can= pd.read_excel('https://ibm.box.com/shared/static/lw190pt9zpy5bd1ptyg2aw15awomz9pu.xlsx',
                 sheet_name='Canada by Citizenship',
                 skiprows = range(20),
                 skipfooter = 2)
df_can.head()
```

```python
#to see the number of entries
print(df_can.shape)
```

```python
#to clean the data
df_can.drop(['Type','Coverage','AREA','REG','DEV'], axis=1, inplace= True)
df_can.head()
```

```python
#to reame some columns
df_can.rename(columns = {'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace = True)
df_can.head()
```

```python
#to check if the column headers are of type string
all(isinstance(column,str) for column in df_can.columns)
```

```python
#to convert all column headers to type string
df_can.columns = list(map(str,df_can.columns))
```

```python
all(isinstance(column,str) for column in df_can.columns)
```

```python
#to set country name as index
df_can.set_index('Country', inplace= True )
df_can.head()
```

```python
#to remove the country label
df_can.index.name = None
df_can.head()
```

```python
#to add a total column
df_can['Total'] = df_can.sum(axis=1)
df_can.head()
```

```python
#to create a list of years from 1980 to 2013
years = list(map(str, range(1980,2014)))
```

```python
#to visulaize data using matplotlib
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
print('Matplotlib version: ', mpl.__version__)
```

1. AREA PLOT

```python
df_can.sort_values(['Total'], ascending= False, axis=0,inplace = True) #to sort in descending order
df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',alpha=0.25, stacked= False,figsize = (20,10))
plt.title('Immigration data to Canada from top 5 countries ')
plt.xlabel('years')
plt.ylabel('number of immigrants')
plt.show()
```

2.HISTOGRAM

```python
#to view data of say 2013 data
df_can['2013'].head()
```

```python
#histogram returns 2 values
count, bin_edges = np.histogram(df_can['2013'])
print(count)
print(bin_edges)
```

```python
#to visualize histogram
df_can['2013'].plot(kind='hist',figsize= (8,5), xticks = bin_edges)
plt.title('immigration to canada in 2013')
plt.xlabel('Immigrants')
plt.ylabel('number of countries')
plt.show()
```

3. BAR CHARTS

```python
#to get the data
df_iceland=df_can.loc['Iceland', years]
df_iceland.head()
```

```python
#to plot the data
df_iceland.plot(kind='bar', figsize= (10,6))
plt.title('immigration data of iceland to canada')
plt.xlabel('years')
plt.ylabel('number of immigrants')
plt.show()
```

```python
#to annotate the plot
df_iceland.plot(kind='bar', figsize = (10,6))
plt.title('immigration data of iceland to canada')
plt.xlabel('years')
plt.ylabel('number of immigrants')
plt.annotate('', xy=(32,70), xytext=(28,20), xycoords='data',
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3',color='blue',lw=2)
            )
plt.annotate('2008 - 2011 Financial Crisis', # text to display
             xy=(28, 30),                    # start the text at at point (year 2008 , pop 30)
             rotation=72.5,                  # based on trial and error to match the arrow
             va='bottom',                    # want the text to be vertically 'bottom' aligned
             ha='left',                      # want the text to be horizontally 'left' algned.
            )
plt.show()
```
