
TO PLOT LINE PLOT

```python
#first extract data
years = list(map(str,(2005,2006,2006,2007,2008,2009,2010)))
Albania = df_can.loc['Albania',years]
```

```python
#to see the Albania data
Albania.head()
```

```python
#to plot the Albania data
Albania.index = Albania.index.map(int)
Albania.plot(kind='line')
plt.title('Immigration of Albania between 2005 and 2010')
plt.xlabel('years')
plt.ylabel('number of immigrants')
plt.show()
```
